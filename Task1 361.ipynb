Шаг 2. Реализация алгоритма градиентного спуска.

Необходимо создать новые коэфициенты $((k^{new}_1, k^{new}_0, b^{new})$ типа tf.Variable.

$$
\left\{\begin{array}{l}

\end{array}\right.
$$
, где $loss = \frac{1}{N}\sum\limits^N_{i = 1} (y - \hat{y})^2 $.

$\eta_i$ - шаг сходимости $i$-го коэфициента.

$loss$ - функция потерь, которую стараемся минимизировать.

$\hat{y} = k1^{new} * \overline{x}^2 + \overline{x} * k0^{new} + b^{new}$

## 2

Реализовать алгоритм градиентного спуска, минимизирующего квадрат ошибки.

Шаг 1. Подготовка искусственной выборки для проверки работы алгоритма.

Создать тензор формы (1000,), состоящий из равномерно распределенных случайных величин на отрезке [min_value, max_value]. ($\overline{x}$)

Создать тензор формы (1000,), состоящий из нормально распределенных случайных величин с параметром отклонения $\sigma$, равным .($\overline{N}$)

Задать параметры для построения модели линейной регрессии $(k, b)$.

Модель регрессии должна представлять следующий вид: 
$$
\overline{y} = \overline{x} * k + b + \overline{N}
$$

Изобразить на графике зависимость $\overline{y}$ от $\overline{x}$, используя пакет matplotlib.pyplot.

Шаг 2. Реализация алгоритма градиентного спуска.

Необходимо создать новые коэфициенты $(k^{new}, b^{new})$ типа tf.Variable.
Необходимо задать шаг сходимости.(Можно для каждого коэфициента выбрать свой шаг.)
Сам алгоритм выглядит следующим образом:
$$
\left\{\begin{array}{l}
\hat{k}^{new}_n=\hat{k}^{new}_{n-1}-\eta \cdot \frac{\partial loss}{\partial \hat{k}^{new}} \\
\hat{b}^{new}_n=\hat{b}^{new}_{n-1}-\eta \cdot \frac{\partial loss}{\partial \hat{b}^{new}}
\end{array}\right.$$
, где $loss = \frac{1}{N}\sum\limits^N_{i = 1} (y - \hat{y})^2 $.

$\eta$ - шаг сходимости.

$loss$ - функция потерь, которую стараемся минимизировать.

$\hat{y} = \hat{k}x + \hat{b}$

Для реализации необходимо распространить алгоритм на третий коэфициент, задать количество итераций, шаг сходимости.
Необходимо использовать градиент из пакета Tensorflow.

Также для отображения результата, необходимо вывести получившиеся значения $\hat{k}, \hat{b}$ (должны быть близки к задаваемым) и получившуюся прямую $y = \hat{k}x + \hat{b}$.


В данной лабораторной работе мы реализовали алгоритм градиентного спуска, который используется для нахождения общей тенденции изменения значения функции. Алгоритм используется в машинном обучении для повышения точности и минимизации ошибок. Одно из простейших применений - найти зависимость известного типа между произвольными данными путём (фактически) перебора коэффициентов. Также можно искать минимумы и максимумы функций.
Уменьшение шагов сходимости позволяет увеличить точность подбора коэффициентов, но увеличивает число итераций алгоритма. Вместо фиксированного числа итераций обычно применяется критерий останова. Фиксированный шаг также может заменяться формулой для поиска оптимального шага (метод наискорейшего спуска).
Градиентный спуск в машинном обучении — метод нахождения минимального значения функции потерь для большого количества различных задач машинного обучения.
Метод градиентного спуска с некоторой модификацией широко используется для обучения персептрона и глубоких нейронных сетей, и известен как метод обратного распространения ошибки.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

tf1 = tf.random.uniform((1000,),-2,2) # горизонтальный разброс
tf2 = tf.random.normal((1000,),0,2) # вертикальный разброс

#k1:  4
#k0: 2
#b:  3

k1 = tf.Variable(float(input("k1: ")))
k0 = tf.Variable(float(input("k0: ")))
b_ = tf.Variable(float(input("b: ")))

y = k1 *(tf1*tf1) + k0*tf1 + b_ + tf2
# часть 1 вывод на экран
plt.scatter(tf1, y, 1)
plt.show()

#создание обучаемых переменных
k1 = tf.Variable(0.0)
k0 = tf.Variable(0.0)
b = tf.Variable(0.0)

#кол-во итераций и шаги сходимости
povtor = 1000

## направление
n1 = 0.001
n2 = 0.0015
n3 = 0.5

#алгоритм градиентного спуска
for n in range(povtor):
    # Используем tf.GradientTape для автоматической дифференциации, в этом случае вычисления градиента.
    # Градиент отражает направление роста функции потерь между исходной функцией y и текущей функцией f. Чем больше значение фунции потерь - тем дальше мы от истины.
    # Поэтому двигаясь по антиградиенту k1, k0, b мы можем минимизировать потери => приблизить параметры k1, k0, b к исходной квадратичной функции.
    with tf.GradientTape() as t:
        # Прямое прохождение градиента
        f = k1 * tf1 * tf1 + k0 * tf1 + b
        #функция потерь - Средняя квадратическая ошибка (MSE)
        loss = tf.reduce_mean(tf.square(y - f))
    # Вычисление градиента на обратном прохождении
    dk1, dk0, db = t.gradient(loss, [k1, k0, b])
    # прибавляем антиградиент (вычитаем градиент) с постоянным шагом
    k1.assign_sub(n1 * dk1)
    k0.assign_sub(n2 * dk0)
    b.assign_sub(n3 * db)
#часть 2 вывод и построение графика
func = k1 *(tf1*tf1) + k0*tf1 + b
plt.scatter(tf1, func, 1)
plt.show()

# Как работает градиент и вывод расширить (как понимаю задачу, что она делает, введение различных лернинг рейтов, количество операций)
```
